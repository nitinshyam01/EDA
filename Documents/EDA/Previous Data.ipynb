## Defaulted

```python
df_0[(df_0.DAYS_LAST_DUE_1ST_VERSION<0)&(df_0.DAYS_LAST_DUE==365243)].shape[0]
```


DAYS_FIRST_DUE is the due date of first instalement in days relative to current application, if the first due date is after current application day its value =365243.  
DAYS_LAST_DUE_1ST_VERSION  =365243 for card users, in other cases it is the no of days wrt current application when last due of the previous application is supposed to be paid.  
DAYS_LAST_DUE is days relative to current application when the customer paid the last payment , else it is =365243.  
DAYS_TERMINATION is days relative to current application when the contract was terminated, else=365243. Generally one week ore so after last payment.

```python
# convert data in days to months
for i in ["DAYS_DECISION","DAYS_FIRST_DRAWING","DAYS_FIRST_DUE","DAYS_LAST_DUE_1ST_VERSION","DAYS_LAST_DUE","DAYS_TERMINATION"]:
    df_0["Months"+i[4::]]=df_0.apply(lambda x: round(x[i]/30,1),axis=1)
    

```

### the diffrence between columns  Months_LAST_DUE and Months_LAST_DUE_1ST_VERSION gives the months defaulted in last payment.


```python
#lets create column Default_last_in_month
df_0["late_payment_last_in_month"]=df_0.apply(lambda x:round((x.Months_LAST_DUE-x.Months_LAST_DUE_1ST_VERSION),1)  if ((x.Months_LAST_DUE-x.Months_LAST_DUE_1ST_VERSION)>0)&(x.Months_LAST_DUE!=round(365243/30,1)) else (0 if (x.Months_LAST_DUE-x.Months_LAST_DUE_1ST_VERSION<=0)|(x.Months_LAST_DUE==round(365243/30,1)) else np.nan) ,axis=1)
```
